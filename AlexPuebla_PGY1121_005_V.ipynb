```python
import datetime

departamentos_disponibles = {
    "A": [True] * 10,
    "B": [True] * 10,
    "C": [True] * 10,
    "D": [True] * 10,
}

departamentos_vendidos = {
    "A": [False] * 10,
    "B": [False] * 10,
    "C": [False] * 10,
    "D": [False] * 10,
}

compradores = {}

precios_venta = {
    "A": 3800,
    "B": 3000,
    "C": 2800,
    "D": 3500,
}

def validar_run(run):
    run = run.replace(".", "").replace("-", "")
    if not run[:-1].isdigit() or not run[-1].isdigit():
        return False

    suma = 0
    multiplicador = 2

    for i in range(len(run) - 2, -1, -1):
        suma += int(run[i]) * multiplicador
        multiplicador += 1
        if multiplicador == 8:
            multiplicador = 2

    digito_verificador = 11 - (suma % 11)
    if digito_verificador == 11:
        digito_verificador = 0
    elif digito_verificador == 10:
        return False

    return digito_verificador == int(run[-1])
def mostrar_menu():
    print("|=======================================================|")
    print("|\t\t Inmobiliaria CASA FELIZ \t\t\t|")
    print("|=======================================================|")
    print("| 1. Comprar departamento o arrendar departamento\t|")
    print("| 2. Mostrar departamentos disponibles\t\t\t|")
    print("| 3. Ver listado de compradores\t\t\t\t|")
    print("| 4. Mostrar ganancias totales\t\t\t\t|")
    print("| 5. Salir\t\t\t\t\t\t|")
    print("|=======================================================|")
    opcion = input("Seleccione una opción: ")
    return opcion

def comprar_arrendar_departamento():

    rut = input("Ingrese el RUT: ")
    rut = rut.replace("-", "").replace(".", "")
    if not rut[:-1].isdigit() or rut[-1] not in "0123456789Kk" or len(rut) < 2:
        print("RUT inválido")
    elif not validar_run(rut):
        print("Dígito verificador incorrecto para el RUT ingresado")
    elif len(rut) < 9:
        print("El RUT debe tener al menos 9 dígitos, incluido su dígito verificador")
    else:
        departamento_valido = False
        while not departamento_valido:
            try:
                piso = int(input("Ingrese el número de piso (1-10): "))
                tipo = input("Ingrese el tipo de departamento (A, B, C, D): ").upper()

                if tipo in departamentos_disponibles.keys() and 1 <= piso <= 10:
                    if departamentos_disponibles[tipo][piso - 1]:
                        departamento_valido = True
                    else:
                        print("El departamento seleccionado no está disponible.")
                else:
                    print("El tipo de departamento o el número de piso ingresado es inválido.")
            except ValueError:
                print("El número de piso ingresado es inválido. Por favor, ingrese un número entero del 1 al 10.")

        if departamento_valido:
            opcion = input("Seleccione una opción (C para confirmar compra o A para anular compra: ").upper()
            while opcion not in ["C", "A"]:
                print("La opción ingresada es inválida.")
                opcion = input("Seleccione una opción válida (C para comprar, A para arrendar): ").upper()

            if opcion == "C":
                if tipo in departamentos_vendidos.keys():
                    if departamentos_vendidos[tipo][piso - 1]:
                        print("El departamento seleccionado ya está vendido.")
                    else:
                        departamentos_vendidos[tipo][piso - 1] = True
                        departamentos_disponibles[tipo][piso - 1] = False
                        compradores[(tipo, piso)] = rut
                        print("Operación realizada correctamente. El departamento ha sido vendido.")
            elif opcion == "A":
                    print("Volviendo al menu inicial")
                    main()

def mostrar_departamentos_disponibles():
    print("\nEstado de los departamentos:")
    print("---------------------------")
    print("|     \t| TIPO DEPARTAMENTO |")
    print("| PISO\t| A | B | C | D |")
    for piso in range(10, 0, -1):
        print("| ", piso, end="  \t|")
        for tipo in ["A", "B", "C", "D"]:
            if departamentos_disponibles[tipo][piso - 1]:
                print("___", end="|")
            elif departamentos_vendidos[tipo][piso - 1]:
                print("_X_", end="|")
            else:
                print("___", end="|")
        print()
    print()


def mostrar_listado_compradores():
    print("\nListado de compradores:")
    print("-----------------------")
    sorted_compradores = sorted(compradores.items(), key=lambda x: x[1])
    for (tipo, piso), run in sorted_compradores:
        print(f"{run} - {tipo}{piso}")
    print()

def mostrar_ganancias_totales():
    total_vendidos = sum(sum(departamentos_vendidos[tipo]) for tipo in departamentos_vendidos.keys())

    print("\nVentas totales:")
    print("----------------")
    print("| TIPO DE DEPARTAMENTO  | CANTIDAD \t|  TOTAL      |")
    total_ventas = 0
    for tipo in departamentos_disponibles.keys():
        cantidad_vendidos = sum(departamentos_vendidos[tipo])
        total_vendidos_tipo = cantidad_vendidos * precios_venta[tipo]
        total_tipo = total_vendidos_tipo
        print(f"| TIPO {tipo} {precios_venta[tipo]:^6} UF \t| {cantidad_vendidos:^8} \t| {total_tipo:>8} UF |")
        total_ventas += total_tipo

    print("|        TOTAL          |", f"{total_vendidos:^8} \t| {total_ventas:>8} UF |")
    print()




def main():
    print("\tBienvenido a la inmobiliaria 'CASA FELIZ'")

    while True:
        opcion = mostrar_menu()

        if opcion == "1":
            comprar_arrendar_departamento()
        elif opcion == "2":
            mostrar_departamentos_disponibles()
        elif opcion == "3":
            mostrar_listado_compradores()
        elif opcion == "4":
            mostrar_ganancias_totales()
        elif opcion == "5":
            nombre = input("Ingrese su nombre: ")
            apellido = input("Ingrese su apellido: ")
            fecha_actual = datetime.date.today()
            print("\nGracias por utilizar el sistema,", end=" ")
            if nombre.isalpha():
                print(nombre, end=" ")
            else:
                print("señor/a", end=" ")
            if apellido.isalpha():
                print(apellido + ".\n")
            else:
                print("Estamos para servirle.\n")
            print("Fecha de salida:", fecha_actual)
            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")

if __name__ == "__main__":
    main()
```
